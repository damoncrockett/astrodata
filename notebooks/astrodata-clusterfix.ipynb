#### similarity

```python
X = d.iloc[:,1:]
```

```python
class cluster:
    dismissed = []
    chosen = []
    
    def __init__(self, matrix, exemplar):
        self.matrix = matrix
        self.field = list(matrix.index)
        self.field.remove(exemplar)
        self.exemplar = exemplar
```

```python
def nn(cluster,n=10):
    tmp = cluster.matrix.loc[cluster.field]
    distances = []
    for i in tmp.index:
        distances.append(np.linalg.norm(cluster.matrix.loc[exemplar]-tmp.loc[i]))
    tmp['distances'] = distances
    tmp = tmp.sort_values(by='distances')
    nearest_neighbors = list(tmp.index[:n])
    return nearest_neighbors 
```

```python
moons = cluster(X,exemplar)
```

```python
neighbors = nn(moons,n=100)
```

```python
dv.montage(d.local_path.loc[neighbors],d.loc[neighbors],'local_path',idx=True,thumb=64)
```

```python
def cull(cluster,neighbors,dismiss):
    cluster.dismissed = cluster.dismissed + dismiss 
    diff = [item for item in neighbors if item not in dismiss]
    cluster.chosen = cluster.chosen + diff
    cluster.field = [item for item in cluster.field if item not in neighbors]
```

```python
def keep(cluster,neighbors,keep):
    cluster.chosen = cluster.chosen + keep 
    diff = [item for item in neighbors if item not in keep]
    cluster.dismissed = cluster.dismissed + diff
    cluster.field = [item for item in cluster.field if item not in neighbors]
```

```python
cull(moons,neighbors,[])
```
